```python
#drop null
df = df.dropna(subset=["year", "danceability", "energy"])
```

```python
#checking for duplicate values
df.duplicated().value_counts()
```

```python
df = df.drop_duplicates(subset=["song", "artist", "year"])
```

```python
df = df[(df["year"] >= 1960) & (df["year"] <= 2024)]
```

```python
df["decade"] = (df["year"] // 10) * 10
df["duration_min"] = df["duration_ms"] / 60000

df["artist"] = df["artist"].str.strip().str.lower()
# Add energy_level column
df["energy_level"] = pd.cut(df["energy"], bins=[0, 0.4, 0.7, 1.0],labels=["Low", "Medium", "High"])
# Normalize genre
df["genre"] = df["genre"].str.lower().str.replace("-", " ").str.strip()

df["genre_main"] = df["genre"].apply(lambda x: x.split(",")[0].strip())
df['genre_main'] = df['genre_main'].replace('set()', 'unknown')
```

```python
#top artist
artist_df = df[['artist', 'popularity']].groupby('artist').mean().sort_values(by='artist')
artists = df['artist'].value_counts().sort_index()
artist_df['total songs'] = artists.values

artist_df.sort_values(by='total songs',ascending=False, inplace=True)

artist_df.reset_index(inplace=True)
top10 = artist_df[:10]
top10
```

```python
top_genres = df["genre_main"].value_counts().head(5).index
print(top_genres)

df_top = df[df["genre_main"].isin(top_genres)]
genre_group = df_top.groupby(['decade', 'genre_main']).size().reset_index(name='count')
genre_group
```

```python
genre_pivot = genre_group.pivot(index='genre_main', columns='decade', values='count')
genre_pivot = genre_pivot.fillna(0)
genre_pivot
```
